### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Drop rows with missing horsepower for this task
df_clean = df.dropna(subset=["horsepower"])

# Compute the median mpg
median_mpg = df_clean["mpg"].median()

# Filter cars with mpg above the median
above_median_mpg = df_clean[df_clean["mpg"] > median_mpg].copy()

# Compute horsepower-to-weight ratio
above_median_mpg["hp_to_weight"] = above_median_mpg["horsepower"] / above_median_mpg["weight"]

# Find the max ratio
max_ratio = above_median_mpg["hp_to_weight"].max()

# Get the car(s) with that max ratio
best_ratio_cars = above_median_mpg[above_median_mpg["hp_to_weight"] == max_ratio]

best_ratio_cars[["name", "mpg", "horsepower", "weight", "hp_to_weight"]]
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute average mpg by year and origin
avg_mpg_by_year_origin = df.groupby(["model_year", "origin"])["mpg"].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x="model_year", y="mpg", hue="origin", marker="o")
plt.title("Average MPG Over the Years by Car Origin")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
# Drop rows with missing horsepower again
scatter_data = df.dropna(subset=["horsepower"]).copy()

# Define hue order
hue_order = ['japan', 'europe', 'usa']

# Plot
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(
    data=scatter_data,
    x="horsepower",
    y="weight",
    hue="origin",
    size="mpg",
    palette="Set1",
    hue_order=hue_order,
    alpha=0.7
)

plt.title("Horsepower vs. Weight of Cars\n(Size = MPG, Color = Origin)")
plt.xlabel("Horsepower")
plt.ylabel("Weight")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Group by car name and count appearances and std dev of mpg
car_stats = df.groupby("name").agg(
    count=("model_year", "nunique"),
    mpg_std=("mpg", "std"),
    avg_mpg=("mpg", "mean")
).reset_index()

# Filter for consistent cars (more than 1 year, std < 1.0)
consistent_cars = car_stats[(car_stats["count"] > 1) & (car_stats["mpg_std"] < 1.0)]

# Sort by count descending, then avg_mpg descending
consistent_cars_sorted = consistent_cars.sort_values(by=["count", "avg_mpg"], ascending=[False, False])

consistent_cars_sorted.reset_index(drop=True)
```
